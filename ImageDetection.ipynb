Usando la RetinaNet invece, con lo strsso codice, ci sono problemi portando alla presenza di fin troppe boxes (molte delle quali collocate male, identificando oggetti che non corrispondono). Ho provato a modificare il valore "box_score_thresh=0.9" nel blocco 27, ma non sembra portare effetto.

```python
from torchvision.io.image import read_image
#from torchvision.models.detection import retinanet_resnet50_fpn_v2, RetinaNet_ResNet50_FPN_V2_Weights
#from torchvision.models.detection import retinanet_resnet50_fpn, RetinaNet_ResNet50_FPN_Weights
from torchvision.models.detection import fasterrcnn_resnet50_fpn_v2, FasterRCNN_ResNet50_FPN_V2_Weights
from torchvision.utils import draw_bounding_boxes
from torchvision.transforms.functional import to_pil_image
```

```python
img = read_image("img_ID/One_day.jpg")
```

```python
#pt_weights = RetinaNet_ResNet50_FPN_V2_Weights(weights='DEFAULT', num_classes=3)
#rt_model = retinanet_resnet50_fpn_v2(weights=pt_weights, num_classes=3, box_score_thresh=0.9)

#pt_weights = RetinaNet_ResNet50_FPN_Weights.DEFAULT
#rt_model = retinanet_resnet50_fpn(weights=pt_weights, box_score_thresh=0.9)

pt_weights = FasterRCNN_ResNet50_FPN_V2_Weights.DEFAULT
rt_model = fasterrcnn_resnet50_fpn_v2(weights=pt_weights, box_score_thresh=0.9)

rt_model.eval()
```

```python
preprocess = pt_weights.transforms()
```

```python
batch = [preprocess(img)]
```

```python
prediction = rt_model(batch)[0]
labels = [pt_weights.meta["categories"][i] for i in prediction["labels"]]
box = draw_bounding_boxes(img, boxes=prediction["boxes"],
                          labels=labels,
                          colors="red",
                          width=4, font_size=30
                         )
im = to_pil_image(box.detach())
im.show()
```
